## Step 1: Run Inference on Pre-trained Gemma Model
* Load the model
* Load the tokenizer
* Visualize the model architecture
* Finally, query the model with a prompt and see the response

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "google/gemma-2b"
# model_name = "meta-llama/Llama-2-7b-hf"

# Load the model
model = AutoModelForCausalLM.from_pretrained(model_name,
                                            #  torch_dtype=torch.float16
                                             )

# Load the autotokenize. Auto: Based on the model name it loads the tokenizer corresponding to that model. Taken care off by class
tokenizer = AutoTokenizer.from_pretrained(model_name,
                                          # torch_dtype=torch.float16
                                          )
print(model)
```

## Step 3: Finetune using Parameter Efficient Finetuning (PEFT)
- Create your own dataset
  - Visualizing the dataset
  - Cleaning and Preprocessing
  - Uploading the created dataset to HF hub
- Fine-tune Gemma 2B model with the dataset

```python
import torch
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    HfArgumentParser,
    TrainingArguments,
    logging,
)
from peft import LoraConfig, PeftModel
from trl import SFTTrainer
```

**Create and load our own Dataset**

```python
"""
We don't have any in-house data so we use the databricks dataset but we modify it and
create a mini dataset for fine-tuning
"""

from datasets import load_dataset

dataset_name = "databricks/databricks-dolly-15k"
dataset = load_dataset(dataset_name, split="train")
```

```python
"""
We first get a count/number of records of the different categories in the dataset
"""

from collections import defaultdict

categries_count = defaultdict(int)

for __, data in enumerate(dataset):
    categries_count[data['category']] += 1

print(categries_count)
```

```python
"""
To create our own mini dataset, we filter out the records that do not have any context
"""

filtered_dataset = []

for __, data in enumerate(dataset):
    # if there is context (just by choice), then continue without adding it to our mini dataset
    if data["context"]:
        continue
    else:
        # If there is no context, we format the data specific to Gemma
        # the Gemma model likes the input to be formatted as follows, with all the Instruction, Response and new lines
        text = f"Instruction:\n{data['instruction']}\n\nResponse:\n{data['response']}"
        filtered_dataset.append({"text": text})

print(filtered_dataset[0:2])
```

```python
"""
Create a jsonl file using the jsonlines lib

Dump the dataset to the jsonl file and upload it

We can create a new dataset on HuggingFace Hub

The data format is different than the original dataset (see video)
- instruction and response is combined as one feaure
"""

# convert to json and save the filtered dataset as jsonl file
import jsonlines as jl
with jl.open('dolly-mini-train.jsonl', 'w') as writer:
    writer.write_all(filtered_dataset[0:])
```

## Step 4: Define all the parameters
- LoRA parameters
- bitsandbytes parameters
- training arguments
- Supervised fine-tuning (SFT) parameters



```python
# define some variables - model names
model_name = "google/gemma-2b"
new_model = "gemma-ft" # give a new model name for our fine-tuned model

################################################################################
# LoRA parameters. Here, we define 3 main parameters
################################################################################
# LoRA attention dimension
# lora_r = 64 # if we have sufficient memory. reduce to save memory
lora_r = 4 # rank
# Alpha parameter for LoRA scaling
lora_alpha = 16
# Dropout probability for LoRA layers
lora_dropout = 0.1

################################################################################
# bitsandbytes parameters
################################################################################
# Activate 4-bit precision base model loading
use_4bit = True
# Compute dtype for 4-bit base models
bnb_4bit_compute_dtype = "float16"
# Quantization type (fp4 or nf4)
bnb_4bit_quant_type = "nf4"
# Activate nested quantization for 4-bit base models (double quantization)
use_nested_quant = False

################################################################################
# TrainingArguments parameters
################################################################################
# Output directory where the model predictions and checkpoints will be stored and training logs
output_dir = "./results"
# Number of training epochs - go through entire training set once
num_train_epochs = 1
# Enable fp16/bf16 training (set bf16 to True with an A100), this depends on the hardware
fp16 = False
bf16 = False
# Batch size per GPU for training
per_device_train_batch_size = 4 # depends on hardware
# Batch size per GPU for evaluation
per_device_eval_batch_size = 4
# Number of update steps to accumulate the gradients for
gradient_accumulation_steps = 1
# Enable gradient checkpointing
gradient_checkpointing = True
# Maximum gradient normal (gradient clipping)
max_grad_norm = 0.3
# Initial learning rate (AdamW optimizer)
learning_rate = 2e-4
# Weight decay to apply to all layers except bias/LayerNorm weights
weight_decay = 0.001
# Optimizer to use
optim = "paged_adamw_32bit"
# Learning rate schedule (constant a bit better than cosine)
lr_scheduler_type = "constant"
# Number of training steps (overrides num_train_epochs)
max_steps = -1
# Ratio of steps for a linear warmup (from 0 to learning rate)
warmup_ratio = 0.03
# Group sequences into batches with same length
# Saves memory and speeds up training considerably
group_by_length = True
# Save checkpoint every X updates steps
save_steps = 25
# Log every X updates steps
logging_steps = 25

################################################################################
# SFT parameters
################################################################################
# Maximum sequence length to use
max_seq_length = 40 # None # if we set to 128, we need to make sure we have enough compute. 40 seems to work here
# Pack multiple short examples in the same input sequence to increase efficiency
packing = True # False
# Load the entire model on the GPU 0
# device_map = {"": 0}
device_map="auto" # ensures that the training make sures of 2 GPUs if there are
```

```python
"""
Load the quantized model

Set some configs in the BitsAndBytes library
"""

# Load QLoRA configuration
compute_dtype = getattr(torch, bnb_4bit_compute_dtype)

bnb_config = BitsAndBytesConfig(
    load_in_4bit=use_4bit, # Activates 4-bit precision loading
    bnb_4bit_quant_type=bnb_4bit_quant_type, # nf4 normal float 4
    bnb_4bit_compute_dtype=compute_dtype, # float16
    bnb_4bit_use_double_quant=use_nested_quant, # False # specify we are not using nested quant
)
```

```python
# Check GPU compatibility with bfloat16
if compute_dtype == torch.float16 and use_4bit:
    major, _ = torch.cuda.get_device_capability()
    if major >= 8:
        print("Setting BF16 to True")
        bf16 = True
    else:
        bf16 = False
```

## Step 5: Load the Model and Tokenizer

```python
# Load base model
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    token=hf_token, # access to the model we are using
    quantization_config=bnb_config,
    device_map=device_map # make use of all avail GPUs
)
model.config.use_cache = False
model.config.pretraining_tp = 1
# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name,
                                          token=hf_token,
                                          trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right" # Fix weird overflow issue with fp16 training
```

## Step 6: Load the LoRA config

```python
"""
target_modules: there are only a few modules, particularly the attention modules
that we will be fine-tuning

Specify the attention modules that are the target modules for fine tuning our model
"""

# Load LoRA configuration
peft_config = LoraConfig(
    lora_alpha=lora_alpha,
    lora_dropout=lora_dropout,
    r=lora_r,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj","gate_proj", "up_proj"]
)
```

## Step 7: Initialise the Training Arguments

```python
# Set training parameters
training_arguments = TrainingArguments(
    output_dir=output_dir,
    num_train_epochs=num_train_epochs,
    per_device_train_batch_size=per_device_train_batch_size,
    gradient_accumulation_steps=gradient_accumulation_steps,
    optim=optim,
    save_steps=save_steps,
    logging_steps=logging_steps,
    learning_rate=learning_rate,
    weight_decay=weight_decay,
    fp16=fp16,
    bf16=bf16,
    max_grad_norm=max_grad_norm,
    max_steps=max_steps,
    warmup_ratio=warmup_ratio,
    group_by_length=group_by_length,
    lr_scheduler_type=lr_scheduler_type,
    report_to="tensorboard",
)
training_arguments
```

## Step 8: Train the model

Pass the model that we will be fine-tuning to the SupervisedFineTuning Trainer class,

pass the dataset we created

pass the peft configs

```python
# Set supervised fine-tuning parameters and create the trainer object
trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    peft_config=peft_config,
    dataset_text_field="text",
    # formatting_func=format_prompts_fn,
    max_seq_length=max_seq_length,
    tokenizer=tokenizer,
    args=training_arguments,
    packing=packing,
)
```

```python
# Train model
trainer.train()
trainer.model.save_pretrained(new_model)
```

## Step 10: Prompt/Use the newly fine-tuned model

```python
input_text = "What should I do on a trip to Europe?"

"""
We have fine-tuned only the LoRA layers so
we need to merge the base model and adapter layers that have been fine-tuned
"""
base_model = AutoModelForCausalLM.from_pretrained(
    model_name,
    low_cpu_mem_usage=True,
    return_dict=True,
    torch_dtype=torch.float16,
    device_map=device_map,
)

# Merge the base model with the new model we have created
model = PeftModel.from_pretrained(base_model, new_model)
model = model.merge_and_unload()

# Reload tokenizer to save it
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"
```

```python
input_ids = tokenizer(input_text, return_tensors="pt").to("cuda")
print(input_ids)
outputs = model.generate(**input_ids, max_length=128)
print(tokenizer.decode(outputs[0]))
```
